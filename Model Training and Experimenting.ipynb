```python
import numpy as np

def expanding_window_splits(n_samples, initial_train_size, horizon=1, step=1):
    """
    Generate (train_idx, test_idx) for expanding window validation.

    Parameters
    ----------
    n_samples : int
        Total number of rows in the time series (len(df)).
    initial_train_size : int
        Number of observations in the first training window.
    horizon : int, default=1
        Number of observations in each test window (forecast horizon).
    step : int, default=1
        How many observations to move the window forward each iteration.

    Yields
    ------
    train_idx : np.ndarray
        Indices for training set.
    test_idx : np.ndarray
        Indices for test set.
    """
    start_train = 0
    end_train = initial_train_size

    while True:
        start_test = end_train
        end_test = start_test + horizon

        if end_test > n_samples:
            break

        train_idx = np.arange(start_train, end_train)
        test_idx = np.arange(start_test, end_test)

        yield train_idx, test_idx

        # expand training window and move forward
        end_train += step
```

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import xgboost as xgb
```

```python
def run_expanding_cv(model, X, y, initial_train_size, 
                     horizon=1, step=1, metric=mean_absolute_error):
    all_preds = []
    all_true = []

    n_samples = len(y)

    for train_idx, test_idx in expanding_window_splits(
        n_samples=n_samples,
        initial_train_size=initial_train_size,
        horizon=horizon,
        step=step
    ):
        # ðŸ‘‡ use iloc for row-based indexing
        X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
        X_test,  y_test  = X.iloc[test_idx],  y.iloc[test_idx]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        all_preds.extend(np.ravel(y_pred))
        all_true.extend(np.ravel(y_test))

    return metric(all_true, all_preds)
```

```python
models = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.001, max_iter=10000),
    "RandomForest": RandomForestRegressor(
        n_estimators=300,
        random_state=42
    ),
    "GradientBoosting": GradientBoostingRegressor(
        n_estimators=300,
        learning_rate=0.05,
        random_state=42
    ),
    "XGBoost": xgb.XGBRegressor(
        n_estimators=400,
        learning_rate=0.05,
        max_depth=3,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    ),
}
```

```python
initial_train_size = 10   # first 10 years as training
horizon = 1               # predict 1 year ahead each time
step = 1                  # grow training window by 1 year each loop

results = {}

for name, model in models.items():
    mae = run_expanding_cv(
        model,
        X,
        y,
        initial_train_size=initial_train_size,
        horizon=horizon,
        step=step,
        metric=mean_absolute_error
    )
    results[name] = mae
    print(f"{name}: MAE = {mae:.3f}")
```

```python
model_names = list(results.keys())
mae_scores = list(results.values())

plt.figure(figsize=(10, 5))
bars = plt.bar(model_names, mae_scores)

plt.ylabel("MAE (lower is better)")
plt.title(f"Model comparison with expanding-window CV (target = {target_col})")
plt.xticks(rotation=30)

# make bars a bit nicer
for bar in bars:
    bar.set_edgecolor("black")
    bar.set_linewidth(0.7)

plt.tight_layout()
plt.show()
```

```python
from itertools import product
import pandas as pd  # just for pretty results table

def expanding_window_grid_search(
    model_cls,
    param_grid,
    X,
    y,
    initial_train_size,
    horizon=1,
    step=1,
    metric=mean_absolute_error
):
    """
    Custom hyperparameter tuning using expanding-window validation.

    Parameters
    ----------
    model_cls : class
        The model class (e.g., RandomForestRegressor, Ridge, etc.)
        NOT an instance. We'll construct instances inside.
    param_grid : dict
        Dictionary of param_name -> list_of_values, e.g.
        {"n_estimators": [100, 300], "max_depth": [3, 5]}.
    X : pd.DataFrame
        Feature matrix, sorted by time.
    y : pd.Series
        Target vector, sorted by time.
    initial_train_size : int
        Size of first training window.
    horizon : int, default=1
        Forecast horizon (test window size).
    step : int, default=1
        How much to expand the training window each iteration.
    metric : callable, default=mean_absolute_error
        Error function; lower is better.

    Returns
    -------
    best_params : dict
        Hyperparameters with the lowest CV score.
    best_score : float
        The corresponding metric value.
    results_df : pd.DataFrame
        Table of all combinations and their scores.
    """

    # Generate all combinations of hyperparameters
    param_names = list(param_grid.keys())
    param_values = list(param_grid.values())

    combos = list(product(*param_values))

    results = []

    for combo in combos:
        params = dict(zip(param_names, combo))

        model = model_cls(**params)

        score = run_expanding_cv(
            model,
            X,
            y,
            initial_train_size=initial_train_size,
            horizon=horizon,
            step=step,
            metric=metric
        )

        results.append({**params, "score": score})
        print(f"Tested {params} -> score = {score:.4f}")

    results_df = pd.DataFrame(results)
    # lower is better
    best_idx = results_df["score"].idxmin()
    best_row = results_df.loc[best_idx]

    best_params = {k: best_row[k] for k in param_names}
    best_score = best_row["score"]

    return best_params, best_score, results_df
```

```python
param_grid = {
    "n_estimators": [100, 300],
    "max_depth": [3, 5, None],
    "min_samples_leaf": [1, 3]
}
```

```python
initial_train_size = 10
horizon = 1
step = 1

best_params, best_score, results_df = expanding_window_grid_search(
    RandomForestRegressor,
    param_grid,
    X,
    y,
    initial_train_size=initial_train_size,
    horizon=horizon,
    step=step,
    metric=mean_absolute_error
)

print("\nBest params:", best_params)
print("Best score (MAE):", best_score)
display(results_df.sort_values("score"))
```
