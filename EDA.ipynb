```python
from scipy.stats import rankdata, norm

def xicor2(x, y, ties="auto"):
    x = np.asarray(x).flatten()
    y = np.asarray(y).flatten()
    n = len(y)

    if len(x) != n:
        raise IndexError(f"x, y length mismatch: {len(x)}, {len(y)}")

    if ties == "auto":
        ties = len(np.unique(y)) < n
    elif not isinstance(ties, bool):
        raise ValueError(f"expected ties either \"auto\" or boolean, got {ties} ({type(ties)}) instead")
    
    y = y[np.argsort(x)]
    r = rankdata(y, method="ordinal")
    nominator = np.sum(np.abs(np.diff(r)))

    if ties:
        l = rankdata(y, method="max")
        denominator = 2 * np.sum(l * (n - l))
        nominator *= n
    else:
        denominator = np.power(n, 2) - 1
        nominator *= 3

    statistic = 1 - nominator / denominator
    p_value = norm.sf(statistic, scale=2 / 5 / np.sqrt(n))

    return statistic, p_value
```

```python
def plot_and_corr(df, target, y):
    df_clean = df[[target, y]].dropna()
    result = xicor2(df_clean[target], df_clean[y])
    print(target, result)

    corr_value = result[0]  # Extract scalar correlation coefficient

    fig, ax1 = plt.subplots(figsize=(10, 5))
    ax1.plot(df_clean.index, df_clean[target], label=target, color='tab:blue')
    ax1.set_ylabel(target, color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.plot(df_clean.index, df_clean[y], label=y, color='tab:orange')
    ax2.set_ylabel(y, color='tab:orange')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    plt.title(f"{target} vs {y}  XiCorr: {corr_value:.3f}")
    plt.tight_layout()
    plt.show()

    return result
```

```python

#Apply to all columns except the target itself
for c in new_df.columns:
    if c != 'Crime_Index':
        plot_and_corr(new_df.loc['2001':].ffill(), c, 'Crime_Index')
```

Im going to select variables that have a correlation of more than 0.6

```python
select = []
for c in new_df.columns:
    if c != 'Crime_Index':
        result = xicor2(new_df[c], new_df['Crime_Index'])
        if result[0] > 0.6:
            select.append(c)
```
