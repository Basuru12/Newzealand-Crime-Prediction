```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
plt.plot(df['Year'],df['Total'])
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.title('Total Crimes Over Years')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for i in range(len(df.columns)-2):
    plt.plot(df['Year'], df.iloc[:, i+1:17], label=df.columns[i+1])
    
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(6, 3, figsize=(12, 15))
axes = axes.flatten()

for i in range(17):
    axes[i].plot(df['Year'], df.iloc[:, i+1])
    axes[i].set_title(df.columns[i+1])
    axes[i].grid(True)
    
plt.tight_layout()
plt.show()
```
