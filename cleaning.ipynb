```python
import pandas as pd

# assume your dataframe is df

def clean_year(col):
    # if column contains "/", pick only the first year
    if "/" in col:
        return col.split("/")[0]
    return col

df.columns = [clean_year(col) for col in df.columns]
```

```python
df
```

```python
df_long = df.melt(id_vars=df.columns[0], var_name='Year', value_name='Count')
```

```python
df_clean = df_long.pivot(index='Year', columns=df.columns[0], values='Count')
```

```python
df_clean
```

```python
df_clean.columns = [col.split(": ", 1)[-1] if ": " in col else col for col in df_clean.columns]
```
