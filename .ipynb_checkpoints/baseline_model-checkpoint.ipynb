# Naive Forcast Model

```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_dataset.csv')
df.tail()
```

```python
df_data = df.iloc[ :-1, -1]
df_data
```

```python
df_pred = df.iloc[1:, -1]
df_pred
```

```python
from sklearn.metrics import mean_absolute_error

x = mean_absolute_error(df_data,df_pred)
print(x)
```
